####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
sns.set_style('whitegrid')
fig, axes = plt.subplots(1,2, figsize=(12,5))
sns.boxplot(y=sinasc['IDADEMAE'], ax=axes[0], color='pink')
axes[0].set_title('Boxplot da idade das Mães')
axes[0].set_ylabel('Idade')
plt.tight_layout()
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
sinasc.PESO
```

```python
sns.set_style('whitegrid')
fig, axes = plt.subplots(1,2, figsize=(12,5))
sns.boxplot(y=sinasc['PESO'], ax=axes[0], color='skyblue')
axes[0].set_title('Boxplot do Peso dos Bebês')
axes[0].set_ylabel('Peso')
plt.tight_layout()
plt.show()
```

```python
sns.set_style('whitegrid')
fig, axes = plt.subplots(1,2, figsize=(12,5))
sns.boxplot(y=sinasc['IDADEMAE'], ax=axes[0], color='pink')
sns.boxplot(y=sinasc['PESO'], ax=axes[1], color='skyblue')
axes[0].set_title('Boxplot da idade das Mães')
axes[0].set_ylabel('Idade')
axes[1].set_title('Boxplot do Peso dos Bebês')
axes[1].set_ylabel('Peso')
plt.tight_layout()
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
sns.set_style('dark')
fig, axes = plt.subplots(1,2,figsize=(13,5))
sns.histplot(sinasc['IDADEMAE'], bins=20 , kde=True, ax=axes[0], color="Green")
axes[0].set_title('Histograma da Idade das Mães')
axes[0].set_xlabel('Idade')
axes[0].set_ylabel('Frequncia')
plt.tight_layout()
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
sns.set_style('dark')
fig, axes = plt.subplots(1,2,figsize=(12,5))
sns.histplot(sinasc['PESO'], bins=20 , kde=True, ax=axes[0], color="blue")
axes[0].set_title('Histograma do Peso dos Bebês')
axes[0].set_xlabel('Peso')
axes[0].set_ylabel('Quantidade')
plt.tight_layout()
plt.show()
```
